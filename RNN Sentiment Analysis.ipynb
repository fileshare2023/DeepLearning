```python
# build the model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense
```

```python
model = Sequential()
model.add(Embedding(vocabulary_size,32,input_length=500)) # 32 is number of properties to represent words
                                                        # 500 is the number of words in each sentence after padding
model.add(GRU(100))
model.add(Dense(1,activation='sigmoid'))
model.summary()

from livelossplot import PlotLossesKerasTF
from tensorflow.keras.metrics import Precision, Recall

model.compile(loss='binary_crossentropy',optimizer='adam',
              metrics=['accuracy',Precision(), Recall()])
model.fit(x_train_padded,y_train,epochs=15,batch_size=64, validation_data=(x_test_padded,y_test),
         callbacks=[PlotLossesKerasTF()])
```

## clean the data

```python
import re
from nltk.corpus import stopwords
```

```python
text="RT @warrio: I liked her and was happy 868787"
```

```python
def cleanText(text):

    # remove twitter handle and RT
    text_x=re.sub('@[A-Za-z0-9]+',"",text)
    text_x=re.sub('RT',"",text_x)
    # remove special characters and make it lower case
    text_x=re.sub('[^A-Za-z\s]',"",text_x).strip().lower()
    return(text_x)
```

```python
df['cleanedText']=df.text.apply(cleanText)
```

```python
df.head()
```

```python
# encode output variable
import numpy as np
df['sentiment']=np.where(df.sentiment=="Positive",1,0)
```

```python
x = df.cleanedText
y = df.sentiment
```

```python
# tokenize sentences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
vocabulary_size=5000
```

```python
tok = Tokenizer(num_words=vocabulary_size)
```

```python
tok.fit_on_texts(x)
```

```python
tok.word_index
```

```python
# use the indexes to convert text into sequences
```

```python
x[0]
```

```python
x_sequences = tok.texts_to_sequences(x)
```

```python
x_sequences[0]
```

```python
# pad sequences
x_sequences_padded = sequence.pad_sequences(x_sequences,maxlen=500)
```

```python
# split the data
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_sequences_padded,y,random_state=0)
```

```python
model = Sequential()
model.add(Embedding(vocabulary_size,32,input_length=500)) # 32 is number of properties to represent words
                                                        # 500 is the number of words in each sentence after padding
model.add(GRU(100))
model.add(Dense(1,activation='sigmoid'))
model.summary()

model.compile(loss='binary_crossentropy',optimizer='adam',
              metrics=['accuracy',Precision(), Recall()])
model.fit(x_train,y_train,epochs=5,batch_size=64, validation_data=(x_test,y_test),
         callbacks=[PlotLossesKerasTF()])
```
